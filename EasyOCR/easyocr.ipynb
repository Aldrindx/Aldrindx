```python
import easyocr
from PIL import Image
import cv2
```

```python
reader = easyocr.Reader(["en","hi"])   # check documentation for language available
```

```python
Image.open("90432_n.jpg")
```

```python
output1 = reader.readtext("90432_n.jpg")      # detail = 0 , for only text output.
```

```python
# The output will be in a list format, each item represents a bounding box, the text detected and confident level.
output1       
```

```python
for i in output1:
    print(i[1])
```

```python
img = cv2.imread("90432_n.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
spacer = 100
for i in output1: 
    top_left = tuple(i[0][0])
    bottom_right = tuple(i[0][2])
    img = cv2.rectangle(img,top_left,bottom_right,(200,20,25),2)
    spacer+=15
    text = i[1]
    print(text)
    
Image.fromarray(img)
```
