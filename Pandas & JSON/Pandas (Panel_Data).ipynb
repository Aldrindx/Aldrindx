###  json Part -2

```python
with open("blockbusters2.json") as f:
    bb2 = json.load(f)
```

```python
bb2
```

```python
df2 = pd.DataFrame(bb2)
df2
```

```python
records = []
for i in range(len(df2.get("genres"))):
    record = (df2.get("genres").get(str(i)))
    for j in record:
        k = []
        k = (list(j.values()))
        k.append(i)
        records.append(k)
```

```python
records
```

```python
df3 = pd.DataFrame(records,columns = ["genres_id","genres_name","index"])
df3
```

```python
 df2 = df2.reset_index()
```

```python
df2["index"] = df2["index"].astype(int)
```

```python
df2
```

```python
x = pd.merge(df2,df3,on = "index")
```

```python
x
```

```python
del x['genres']
```

```python
x.head()
```

```python
b = pd.DataFrame(dict(x['belongs_to_collection']))
```

```python
b = b.T
b
```

```python
del b['poster_path']
del b['backdrop_path']
```

```python
b = b.reset_index()
b
```

```python
final_df = pd.merge(x,b,on = "index")
```

```python
del final_df["belongs_to_collection"]
```

```python
final_df
```
